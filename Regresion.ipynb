```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

numeric_cols = df.columns

#MinMax deja valores entre [0,1]
scaler = MinMaxScaler()

df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

df
```

# Pregunta 4

```python
from sklearn.model_selection import train_test_split
```

```python
train_data, rest_data = train_test_split(df, test_size=0.4, random_state=42)


test_data, validation_data = train_test_split(rest_data, test_size=0.5, random_state=42)
```

# Pregunta 5

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score
```

```python
X_train, y_train = train_data.drop("Performance Index", axis=1), train_data["Performance Index"]
X_test, y_test = test_data.drop("Performance Index", axis=1), test_data["Performance Index"]
X_val, y_val = validation_data.drop("Performance Index", axis=1), validation_data["Performance Index"]

linear_regression_model = LinearRegression()
linear_regression_model.fit(X_train, y_train)


random_forest_model = RandomForestRegressor(random_state=42)
random_forest_model.fit(X_train, y_train)

decision_tree_model = DecisionTreeRegressor(random_state=42)
decision_tree_model.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "max_depth": [None, 10, 20],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}

rf_model = RandomForestRegressor(random_state=42)
rf_grid_search = GridSearchCV(rf_model, param_grid, cv=5, scoring='neg_mean_squared_error')
rf_grid_search.fit(X_train, y_train)

dt_model = DecisionTreeRegressor(random_state=42)
dt_grid_search = GridSearchCV(dt_model, param_grid, cv=5, scoring='neg_mean_squared_error')
dt_grid_search.fit(X_train, y_train)

best_rf_model = rf_grid_search.best_estimator_
best_dt_model = dt_grid_search.best_estimator_

y_val_rf = best_rf_model.predict(X_val)
y_val_dt = best_dt_model.predict(X_val)
```

#Pregunta 6

```python
import numpy as np
```

```python
models = {
    "Linear Regression": (linear_regression_model, linear_regression_model.predict(X_val)),
    "Random Forest Regressor": (random_forest_model, random_forest_model.predict(X_val)),
    "Decision Tree Regressor": (decision_tree_model, decision_tree_model.predict(X_val)),
    "Random Forest Regressor optimizado": (best_rf_model, best_rf_model.predict(X_val)),
    "Decision Tree Regressor optimizado": (best_dt_model, best_dt_model.predict(X_val))
}

for name, (model, y_pred) in models.items():
    mse = mean_squared_error(y_val, y_pred)
    r2 = r2_score(y_val, y_pred)
    print(f"Modelo: {name}")
    print(f"MSE: {mse}")
    print(f"R^2: {r2}")
    print("-------------------------------------")
```

```python
feature_importances = best_rf_model.feature_importances_
feature_names = X_train.columns

indices = np.argsort(feature_importances)[::-1]

feature_importance_df = pd.DataFrame({"Caracter√≠stica": feature_names[indices], "Importancia": feature_importances[indices]})

feature_importance_df
```
